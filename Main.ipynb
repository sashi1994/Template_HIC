# Plotting of ATD Parts

## Only options available are:

- **Head**
- **Chest**
- **Pelvis**

For each of these ATD parts, we capture comprehensive motion data across the following aspects:

- **Acceleration**: Provides insight into the linear acceleration experienced by each part over time.

- **Angular Velocity**: Reflects the rate of rotation or change in angular position for each ATD part.

- **Coordinates**: Spatial data points that describe the position of each part in terms of x, y, and z axes, allowing us to track their movement and alignment.

### Head ATD
* Acceleration of Head
* Angular velocity of Head

```python
# Call acceleration if there's a valid object mapp
Graphics.plot(Simulation_Test['Time'].timeseries(),[Simulation_Test['ATD_Left'].Acceleration('Head')],title='Acceleration',time_limit=(150,250))
# Call angular velocity if there's a valid object mapp
Graphics.plot(Simulation_Test['Time'].timeseries(),[Simulation_Test['ATD_Left'].Angular_Velocity('Head')],title='Angular_Velocity',time_limit=(150,250))
```
